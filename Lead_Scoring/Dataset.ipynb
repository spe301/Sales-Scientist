## temporary data fillers

* signifigant Refferrals: 0
* total visits: 9,000
* bounce rate, visit duration, and pages per visit: mean
* revenue: (bounce rate / 100)(visits)(pages per visit)(1.29)
* profit: 
* adspend: 
* dominant social: most common platform for businesses with the same offering 
    * information, consulting, or coaching: (revenue(0.9)) - profit
    * software: (revenue(n)) - profit
    * ecommerce: (revenue(n)) - profit

```python
def estimateRevenue(visits, bounceRate, nPages):
    return ((bounceRate/100)*visits)*nPages*1.29
```

```python
r = []
for i in range(len(df)):
    r.append(estimateRevenue(df['page total visits'][i], df['bounce rate'][i], df['pages per visit'][i]))
    
df['revenue'] = r
df.head()
```

```python
def estimateSocial(df, offer):
    temp = dict(df.loc[df['offering'] == offer]['dominant social'].value_counts())
    counts = {}
    for k, v in temp.items():
        counts[v] = k
    try:
        return counts[max(counts)]
    except:
        return 'Facebook'
```

```python
ds = [] 
for i in range(len(df)):
    if type(df['dominant social'][i]) == str:
        ds.append(df['dominant social'][i])
    else:
        ds.append(estimateSocial(df, df['offering'][i]))
        
df['dominant social'] = ds
df.head()
```
